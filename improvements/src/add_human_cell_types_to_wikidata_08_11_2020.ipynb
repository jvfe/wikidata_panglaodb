```python
import pandas as pd

df = pd.read_csv("../results/cell_type_reference_from_panglao_to_wikidata_31_10_2020.csv")
```

```python
alias_en = df["panglao"]

alias_en = ["human " + alias for alias in alias_en]

alias_en = [alias.lower() for alias in alias_en]


label_en = alias_en

replacements = {
    "cells" : "cell",
    "neurons" : "neuron",
    "cytes" : "cyte",
    "lasts" : "last",
    "elets" : "elet",
    "phils" : "phil"
}

for replacement in replacements:
    label_en = [label.replace(replacement, replacements[replacement]) for label in label_en]
    
df["Len"] = label_en
df["Aen"] = alias_en
```

```python
df["Den"] = "cell type of Homo sapiens"

df["Dpt"] = "tipo de célula de Homo sapiens"

df["P361"] = df["wikidata"]

df["P703"] = "Q15978631" # found in human
df["P31"] = "Q189118" # instance of cell type
```

```python
df = df.drop(["wikidata", "panglao"], axis=1)
df = df.dropna()
df.head()
```

```python
df.to_csv("../data/human_cell_types_to_add_08_11_2020.csv", index=False )
```

```python
with open("../quickstatements/add__human_cell_types_08_11_2020.qs", "w+") as qs:

    for i, row in df.iterrows():

        s = "LAST"

        Len = s + "|Len|" + '"' + row["Len"] + '"' + "\n"
        Den = s +"|Den|" + '"' + row["Den"] + '"' + "\n"
        Dpt = s +"|Dpt|" + '"' + row["Dpt"] + '"' + "\n"
        Aen = s +"|Aen|" + '"' + row["Aen"] + '"' + "\n"
        P31 = s + "|P31|" + row["P31"] + "|S248|" + "Q99936939" +   "\n" 
        P703 = s + "|P703|" + row["P703"] + "\n"
        P361 = s + "|P361|" + row["P361"] + "\n"

        qs.write('CREATE \n')
        qs.write(Len)
        qs.write(Den)
        qs.write(Dpt)
        qs.write(Aen)
        qs.write(P31)
        qs.write(P703)
        qs.write(P361)
        qs.write('\n')
```
